```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores.faiss import FAISS
```

```python
from langchain_community.document_loaders import PyMuPDFLoader
```

```python
import os


os.environ['CURL_CA_BUNDLE'] = ''
```

```python
documents = []
for file in os.listdir('/kaggle/input/papers'):
    try:
        loader = PyMuPDFLoader('/kaggle/input/papers' + '/'+ file)
        documents.extend(loader.load_and_split(text_splitter=RecursiveCharacterTextSplitter(chunk_size=900, chunk_overlap=450)))
    except Exception as e:
        print(e)
        
    
```

```python
import torch
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
model_name = "sentence-transformers/paraphrase-multilingual-mpnet-base-v2"
model_kwargs = {'device': device}
encode_kwargs = {'normalize_embeddings': False}
embedding = HuggingFaceEmbeddings(model_name=model_name,
                                  model_kwargs=model_kwargs,
                                  encode_kwargs=encode_kwargs)
```

```python
vector_store = FAISS.from_documents(documents, embedding)
vector_store.save_local('/kaggle/working/')
```

```python
import re

def clean_text_from_pdf(docum):
    for chunk in range(len(docum)):
        text = doc[chunk].page_content
        text = re.sub(r'http[s]?://\S+|www\.\S+', '', text)  
    # Убираем изображения, которые могут быть представлены как ссылки или текст
        text = re.sub(r'\[.*?\]\(.*?\)|(?:download(?:[^\n]*?))|(?:image(?:[^\n]*?))', '', text)
    # Убираем строки, представляющие собой таблицы (например, строка с символами "|")
        text = re.sub(r'(^|\n)\|.*?\|(\n|$)', '', text) # убрать строки с разделителем `|
    # Удаляем специальные символы, оставляя только буквы, цифры и пробелы
        text = re.sub(r'[^a-zA-Zа-яА-Я0-9\s]', '', text)
    # Удаляем лишние пробелы и пробелы в начале и в конце строки
        text = re.sub(r'\s+', ' ', text).strip()
        docum[chunk].page_content = text
    return doc
```

```python
documents = []
```

```python
for file in os.listdir('/kaggle/input/papers'):
    try:
        loader = PDFMinerLoader('/kaggle/input/papers' + '/'+ file)
        doc = loader.load()
        doc = clean_text_from_pdf(doc)
        text_splitter = RecursiveCharacterTextSplitter(chunk_size=900, chunk_overlap=300)
        splitted_data = text_splitter.split_documents(doc)
        documents.extend(splitted_data)        
    except Exception as e:
        print(e)
```

```python
for file in os.listdir('/kaggle/input/paper2'):
    try:
        loader = PDFMinerLoader('/kaggle/input/paper2' + '/'+ file)
        doc = loader.load()
        doc = clean_text_from_pdf(doc)
        text_splitter = RecursiveCharacterTextSplitter(chunk_size=900, chunk_overlap=300)
        splitted_data = text_splitter.split_documents(doc)
        documents.extend(splitted_data)        
    except Exception as e:
        print(e)
```

```python
vector_store_clear = FAISS.from_documents(documents, embedding)
vector_store_clear.save_local('/kaggle/working/')
```
